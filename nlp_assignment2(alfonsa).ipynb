```python
import spacy

nlp = spacy.load("en_core_web_sm")
test_file=open('C:\\Users\\best\\Downloads\\implesys.txt')
data_test=test_file.read()

doc = nlp(data_test)

for ent in doc.ents:
    print(ent.text, ent.start_char, ent.end_char, ent.label_)
```

# Tokenizer

## word tokenization

#### https://spacy.io/usage/linguistic-features#tokenization

# dependency parsersing

#### https://www.reddit.com/r/spacynlp/comments/4n5hda/syntactic_dependency_parsing_usage/
#### https://shirishkadam.com/2016/12/23/dependency-parsing-in-nlp/#:~:text=Syntactic%20Parsing%20or%20Dependency%20Parsing,a%20syntactic%20structure%20to%20it.&text=These%20parse%20trees%20are%20useful,in%20the%20semantic%20analysis%20stage.
#### https://www.youtube.com/watch?v=l307P0uZASg
#### https://realpython.com/natural-language-processing-spacy-python/#:~:text=Remove%20ads-,Dependency%20Parsing%20Using%20spaCy,the%20root%20of%20the%20sentence.

```python
for word in doc:
    if word.dep_ == 'nsubj':
        print('Subject:', word.text)
    if word.dep_ == 'dobj':
        print('Object:', word.text)
```

```python
for token in doc:
    print (token.text, token.tag_, token.head.text, token.dep_)
```
