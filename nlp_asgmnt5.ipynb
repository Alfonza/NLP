- Ensemble methods is a machine learning technique that combines several base models in order to produce one optimal predictive model
- They combine the decisions from multiple models to improve the overall performance

##  Ensemple Methods
1. Basic Ensemble Techniques
####   Max Voting  
     * Multiple models are trained on same dataset
     * Then for prediction, the data is predicted by using all trained models.
     * The maximum voted prediction from all the prediction is taken as the final prediction
####  Averaging
     * Averaging is similer to maxvoting, but instead of taking the maximum voted prediction,it take the average of all model predictions as final prediction 
####  Weighted Average
     * It is an extension of averaging
     * In some cases, we have to give different priority to different model predictions
     * Therefore each model is assigned with a weight value. more prioritiesed models will have greater weight value
     * The final prediction is
\begin{equation*}
final\_prediction\_value   = {\sum_{k=1}^n w_k p_k}
\end{equation*}

## Bagging

- bootstrapping and aggregating



1. The training dataset is converted into multiples bootstrapped subsamples
2. Then each model is trained for each bootstrapped subsamples
3. Then using some algorithm, the results are aggregated to find the best predictor

###  Bootstrapping

- In boostrapping, a subset of training dataset is taken as a data 
- If the probability distribution of bootstrapped dataset is not uniform, the existing data will be added to make it uniform

### How it works:-
- Random forest algorithm randomly selects observations and features to build several decision trees
- During prediction, the data is trained through all the decision trees.
- The average of prediction results of all the decision trees is the final prediction result

### Important hyperparameters of Random forest
- <b>n_estimators</b> = number of decision trees to construct. 
    higher value gives better predictive power. but increases computation cost
- <b>max_features</b> = maximum number of features random forest considers to split a node
- <b>n_jobs</b>       = number of processors to use. giving value -1 means no limit
- <b>oob_score </b>   = Cross validation score

### Advantages of random forest
- Versatility- It can be used for both regression and classification tasks, and it’s also easy to view the relative importance it assigns to the input features.

- Good prediction result.
- Understanding the hyperparameters is easy
- Less chance for overfitting


### Disadvantages of random forest
- Sometimes too slow and ineffective for real-time predictions. 
- Random forest is a predictive modeling tool and not a descriptive tool

# Boosting
#### https://www.analyticsvidhya.com/blog/2015/11/quick-introduction-boosting-algorithms-machine-learning/
#### https://www.geeksforgeeks.org/boosting-in-machine-learning-boosting-and-adaboost/

which attempts to build a strong classifier from the number of weak classifiers. It is done building a model by using weak models in series.
1. A subset is created from the original dataset.
2. Initially, all data points are given equal weights.
3. A base model is created on this subset.
4. This model is used to make predictions on the whole dataset.
5. Errors are calculated using the actual values and predicted values.
6. The observations which are incorrectly predicted, are given higher weights.
7. then another model is made and trained on the updated dataset.
8. this is iterated many times
9. thus errors are recorrected in each iteration and better models are generated.
10. The final model (strong learner) is the weighted mean of all the models (weak learners)

## Decision Tree

#### https://www.geeksforgeeks.org/decision-tree/
#### 
- Decision tree is the most powerful and popular tool for classification and prediction.
- A Decision tree is a flowchart like tree structure, where each internal node denotes a test on an attribute, each branch represents an outcome of the test, and each leaf node (terminal node) holds a class label.



- Find each features information gain. In that which feature have high information gain value it consider the first branch

### Advantages
- Decision trees are able to generate understandable rules.
- Decision trees perform classification without requiring much computation.
- Decision trees are able to handle both continuous and categorical variables.
- Decision trees provide a clear indication of which fields are most important for prediction or classification.


### Disadvantages
- Decision trees are less appropriate for estimation tasks where the goal is to predict the value of a continuous attribute.
- Decision trees are prone to errors in classification problems with many class and relatively small number of training examples.
- Decision tree can be computationally expensive to train. The process of growing a decision tree is computationally expensive. At each node, each candidate splitting field must be sorted before its best split can be found. In some algorithms, combinations of fields are used and a search must be made for optimal combining weights. Pruning algorithms can also be expensive since many candidate sub-trees must be formed and compared.
