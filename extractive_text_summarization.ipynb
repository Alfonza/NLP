## Extractive Text Summarization

### Term Frequency Method

```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
```

```python
document=open('half girl friend.txt','r')
```

```python
data=document.read()
data
```

```python
stopwords=list(STOP_WORDS)
stopwords
```

```python
nlp=spacy.load('en_core_web_sm')
```

```python
docx=nlp(data)
```

```python
punctuation=punctuation + '\n'
punctuation
```

```python
word_freq={}
for word in docx:
    if word.text.lower() not in stopwords:
        if word.text.lower() not in punctuation:
            if word.text not in word_freq.keys():
                word_freq[word.text]=1
            else:
                word_freq[word.text]+=1
```

```python
word_freq
```

```python
max_freq=max(word_freq.values())
max_freq
```

```python
for word in word_freq.keys():
    word_freq[word]=(word_freq[word]/max_freq)
```

```python
word_freq
```

```python
sentence_list=[sentence for sentence in docx.sents]
    
```

```python
sentence_rank={}
for sent in sentence_list:
    for word in sent:
        if word.text.lower()in word_freq.keys():
            if len(sent.text.split(' '))<30:
                if sent not in sentence_rank.keys():
                    sentence_rank[sent]=word_freq[word.text.lower()]
                else:
                    sentence_rank[sent]+= word_freq[word.text.lower()]
```

```python
sentence_rank
```

```python
print(max(sentence_rank.values()))
```

```python
from heapq import nlargest
```

```python
summarize_sent=nlargest(7,sentence_rank,key=sentence_rank.get)
```

```python
summarize_sent
```

```python
final_sent=[word.text for word in summarize_sent]
```

```python
summary=' '.join(final_sent)
```

```python
summary
```

```python
from nltk.stem import PorterStemmer
porter = PorterStemmer()
porter.stem(data)
```

```python
new_doc=[]
for sentence in docx.sents:
    a=[]
    for word in sentence:
        if word.text.lower() not in stopwords:
            if word.text.lower() not in punctuation:
                
                a.append(porter.stem(word.text.lower()))
                a.append(" ")
    l="".join(a)
    new_doc.append(l)
    #new_doc.append(" ")
print(new_doc)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
obj = TfidfVectorizer()
corpus=new_doc
X = obj.fit_transform(corpus)
print(X)
```
